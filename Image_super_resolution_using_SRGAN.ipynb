```
preprocess = T.Compose([
    T.ToTensor(),
    T.Normalize([0.485, 0.456, 0.406],
                [0.229, 0.224, 0.225]),
    T.Lambda(lambda x: x.to(device))
])

postprocess = T.Compose([
    T.Lambda(lambda x: (x.cpu().detach()+1)/2),
])
```

```
image = readPIL('babe.jpg')
image.size
# (260,181)
image = image.resize((130,90))
im = preprocess(image)
```

```
def process_images(img):
    image = readPIL(img)

    resize_im = image.resize((130,90))

    low_res = preprocess(resize_im)

    orig_im = T.ToTensor()(image.resize((520,360)))

    sr = model(low_res[None])[0]
    sr = postprocess(sr)

    return orig_im,resize_im,sr
```

```
subplots([orig_im,image, sr], nc=3, figsize=(10,10), titles=['Original image','Low resolution','High resolution image'])
```

```

orig_im,low_res,sr = process_images('baldie.jpg')
subplots([orig_im,low_res, sr], nc=3, figsize=(10,10), titles=['Original image','Low resolution','High resolution image'])
```
